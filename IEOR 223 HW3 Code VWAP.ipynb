```python
import pandas as pd 
import numpy as np 
import yfinance as yf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
msft= yf.download(tickers = "MSFT",period = "2d",interval = "1m",group_by = 'ticker')
msft= msft[:15]
```

```python

stock=msft
```

```python
stock["Avg Price"]=(stock["High"]+stock["Low"]+stock["Close"])/3
stock["Cum Vol"] = np.cumsum(stock["Volume"])
```

```python
stock["AvgP * Volume"] = (stock["Avg Price"] * stock["Volume"])
stock["Cum AvgP*Vol"]=np.cumsum(stock["AvgP * Volume"])
stock["VWAP"] = stock["Cum AvgP*Vol"]/ stock["Cum Vol"]
```

```python
stock["Comparison"] = (stock["VWAP"]>stock["Avg Price"])
stock
```

```python
x = stock.index
```

```python
#plt.plot(stock["VWAP"])
#plt.plot(stock["Avg Price"])
#plt.legend()

fig, ax = plt.subplots()
ax.plot(x, stock["VWAP"], '--g', label='VWAP')
ax.plot(x, stock["Avg Price"], '-r', label='Avg Price')
leg = ax.legend();
```

From the above, the stock appears underpriced using VWAP strategy at09:38 AM but becomes overpriced after that as Avg P > VWAP.
